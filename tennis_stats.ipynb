```
url = 'https://docs.google.com/spreadsheets/d/1x742LIKrHtTW8faLf8JtDWo6UWT6lbXWHn2l9RpUa4k/edit?ts=5d9f2d00#gid=1238200745' # @param {type: 'string'}

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Call once to configure Bokeh to display plots inline in the notebook.
output_notebook()

from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())

worksheet = gc.open_by_url(url).sheet1

columns = worksheet.row_values(1)
# get_all_values gives a list of rows.
rows = worksheet.get_all_values()[1:]

# Convert to a DataFrame and render.
import pandas as pd
df = pd.DataFrame.from_records(rows, columns = columns)

df['timestamp'] =  pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S')
```

```
# @title Swing speed
from bokeh.models import ColumnDataSource, Legend,DatetimeTickFormatter, Span
from bokeh.palettes import Category20
from datetime import *


by_type = (df[['timestamp', "swing_type", 'swing_speed[km/h]']].groupby("swing_type").rolling("90D", on='timestamp').mean().reset_index())[::100]

p = figure(plot_width=800, plot_height=800, tools=["reset", "save"])
p.xaxis[0].formatter = DatetimeTickFormatter()

values = by_type.pivot(index='timestamp', columns='swing_type', values='swing_speed[km/h]')

for i, x in enumerate(values):
  series = values[x].dropna()
  p.circle(x=series.index.tolist(), y=series.tolist(), legend = x, color=Category20[20][2*i], size = 8)

p.xaxis.axis_label = "date"
p.yaxis.axis_label = "Swing Speed[km/h]"
p.legend.location = "bottom_left"

vline = Span(location=date(2018, 7, 1), dimension='height', line_color='red', line_width=5, line_dash = "dotted")

p.renderers.extend([vline])
show(p)  
```

```
# @title Spin
from bokeh.models import ColumnDataSource, Legend
from bokeh.palettes import Category20


by_type = (df[['timestamp', "swing_type", 'ball_spin']].groupby("swing_type").rolling("90D", on='timestamp').mean().reset_index())[::100]

p = figure(plot_width=800, plot_height=800, tools=["reset", "save"])

values = by_type.pivot(index='timestamp', columns='swing_type', values='ball_spin')

for i, x in enumerate(values):
  series = values[x].dropna()
  p.circle(x=series.index.tolist(), y=series.tolist(), legend = x, color=Category20[20][2*i], size = 8)

p.xaxis.axis_label = "date"
p.yaxis.axis_label = "Ball Spin"
p.xaxis[0].formatter = DatetimeTickFormatter()
p.legend.location = "bottom_left"


vline = Span(location=date(2018, 7, 1), dimension='height', line_color='red', line_width=5, line_dash = "dotted")

p.renderers.extend([vline])

show(p)  
```
